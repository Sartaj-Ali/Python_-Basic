```python
'''Task 1: Write a program that asks the user about the number of values he/she wants to enter. 
Than enter the values as per the required number, calculate its sum and identify the smallest value among them. 
```

```python
num_value = int(input("HOW MANY VALUES DO YOU WANT TO ENTER"))
total_sum= 0
smallest_value = float('inf') 
for i in range(num_value):
      value = float(input(f"Enter value {i+1}: "))
      total_sum += value
      if value < smallest_value:
         smallest_value = value
print(f"Sum of the values: {total_sum}")
print(f"Smallest value: {smallest_value}")
```

```python
''' Write a function factorial that accepts an integer as parameter and returns its factorial. Using the factorial function,
write a program that evaluates the factorials of the integers from 1 to 5. Print the results in tabular format as following.
```

```python
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)
print("Number\tFactorial")
for i in range(1, 10):
    print(f"{i}\t{factorial(i)}")
```

```python
'''Task 4: The greatest common divisor (GCD) of two integers is the largest integer that evenly divides each of the two numbers. Write function gcd that 
returns the greatest common divisor of two integers. Use the gcd function in your program to determine the GCD of the numbers in the sample output:
```

```python
def gcd(a, b):
   while b != 0:
        a, b = b, a % b
   return a
num1 = int(input("Enter the first number: "))
num2 = int(input("Enter the second number: "))
print("GCD of", num1, "and", num2, "is:", gcd(num1, num2))
```

```python
'''Task 5: An integer is said to be prime if it is divisible only by 1 and itself. For example, 2, 3, 5 and 7 are prime, but 4, 6, 8 and 9 are not.
a) Write a function that determines if a number is prime.
b) Use this function in a program that determines and prints all the prime numbers between 1 and 10,000.
```

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

# Use the function to print all prime numbers between 1 and 10,000
for num in range(1, 10001):
    if is_prime(num):
        print(num)
```

```python
#Task 6: Write a code that prints on screen all the 4-digit Armstrong numbers.
```

```python
# Define a function to check if a number is Armstrong
def is_armstrong(n):
    return n == sum(int(digit) ** 4 for digit in str(n))

# Print all 4-digit Armstrong numbers
for num in range(1000, 10000):
    if is_armstrong(num):
        print(num)
```

```python
#Task 11: Use a for loop to print a box like the one below. Allow the user to specify how wide and how high the box should be.
```

```python
width = int(input("How wide should the box be? "))
height = int(input("How high should the box be? "))

for i in range(height):
    if i == 0 or i == height - 1:
        print("*" * width)
    else:
        print("*" + "*" * (width - 2) + "*")
```

```python
#Task 13: Use for loops to print a diamond like the one below. Allow the user to specify how high the diamond should be.
```

```python
height = int(input("How high should the diamond be? "))
for i in range(height):
    print(" " * (height - i - 1) + "*" * (2 * i + 1))
for i in range(height - 2, -1, -1):
    print(" " * (height - i - 1) + "*" * (2 * i + 1))
```

```python
#Task 14: Write a program that prints a giant letter A like the one below. Allow the user to specify how large the letter should be.
```

```python
lines = int(input('Enter The height of A : '))
def print_alphabet_pattern(lines):
    for a in range(lines):
        if a == 0:
            print((lines - a)  *  ' ' + 1 * ' * ')

        elif a == lines // 2:
            print((lines - a - 1) * ' ' + 1 * ' *********' + 2 * ' ' * (a) + 1 * '')
        else:
            print((lines - a - 1) * ' ' + 1 * ' * ' + 2 * ' ' * (a) + 1 * '*')

print_alphabet_pattern(lines)
```
